# Forecasting Using Meta's Prophet Library

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

def prepare_data_for_prophet(df, crime_type):
    """
    Prepare data for Prophet forecasting by aggregating crime counts by date
    """
    # Ensure we have a datetime index
    if not isinstance(df.index, pd.DatetimeIndex):
        df.loc[:, 'Date'] = pd.to_datetime(df.loc[:, 'Date'])

    # Count crimes by date for the specific crime type
    daily_crimes = df[df['Primary Type'] == crime_type].groupby('Date').size().reset_index()
    daily_crimes.columns = ['ds', 'y']

    return daily_crimes

def forecast_crime_type(df, crime_type, periods=30):
    """
    Forecast crime counts for a specific crime type
    """
    # Prepare data
    crime_data = prepare_data_for_prophet(df, crime_type)

    # Initialize and fit Prophet model
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        seasonality_mode='multiplicative'
    )

    model.fit(crime_data)

    # Create future dates dataframe
    future = model.make_future_dataframe(periods=periods)

    # Make predictions
    forecast = model.predict(future)

    return model, forecast

def plot_forecast(model, forecast, crime_type):
    """
    Plot the forecast results
    """
    fig = model.plot(forecast)
    plt.title(f'Forecast for {crime_type}')
    return fig

def forecast_all_crime_types(df, top_n=5, periods=30):
    """
    Forecast all crime types and return the results
    """
    # Get top N most frequent crime types
    top_crimes = df['Primary Type'].value_counts().head(top_n).index.tolist()

    forecasts = {}
    for crime_type in top_crimes:
        model, forecast = forecast_crime_type(df, crime_type, periods)
        forecasts[crime_type] = {
            'model': model,
            'forecast': forecast
        }

    return forecasts

# Example usage
if __name__ == "__main__":
    # Assuming 'crimes' is your DataFrame with columns: date, Primary Type, etc.
    # Make sure to convert the date column to datetime if it's not already
    crimes_2015.loc[:, 'Date'] = pd.to_datetime(crimes_2015.loc[:, 'Date'])

    # Forecast top 5 most common crime types
    forecasts = forecast_all_crime_types(crimes_2015, top_n=5)

    # Print forecasts and create plots
    for crime_type, result in forecasts.items():
        forecast = result['forecast']
        model = result['model']

        print(f"\nForecast for {crime_type}")
        print("Expected number of crimes for next 30 days:")
        print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
              .tail(30)
              .to_string(index=False))

        # Plot the forecast
        plot_forecast(model, forecast, crime_type)
        plt.show()
```

# Forecasting Using SARIMA

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
import warnings
warnings.filterwarnings('ignore')

def prepare_time_series(df, crime_type):
    """
    Prepare time series data for a specific crime type
    """
    # Convert to daily counts
    daily_crimes = df[df['Primary Type'] == crime_type].groupby('Date').size()
    # Ensure the index is datetime and sorted
    daily_crimes.index = pd.to_datetime(daily_crimes.index)
    daily_crimes = daily_crimes.sort_index()
    # Fill missing dates with zeros
    idx = pd.date_range(daily_crimes.index.min(), daily_crimes.index.max(), freq='D')
    daily_crimes = daily_crimes.reindex(idx, fill_value=0)

    return daily_crimes

def train_test_split_time(series, test_size=30):
    """
    Split time series into train and test sets
    """
    return series[:-test_size], series[-test_size:]

def fit_sarima(train_data, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7)):
    """
    Fit SARIMA model with specified parameters
    """
    model = SARIMAX(
        train_data,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )

    fitted_model = model.fit(disp=False)
    return fitted_model

def forecast_crime_type(df, crime_type, forecast_periods=30):
    """
    Forecast crime counts for a specific crime type
    """
    # Prepare data
    time_series = prepare_time_series(df, crime_type)

    # Split data
    train_data, test_data = train_test_split_time(time_series, test_size=forecast_periods)

    # Fit model
    model = fit_sarima(train_data)

    # Make forecast
    forecast = model.forecast(steps=forecast_periods)

    # Calculate error metrics if we have test data
    mae = mean_absolute_error(test_data, forecast)
    rmse = np.sqrt(mean_squared_error(test_data, forecast))

    return {
        'forecast': forecast,
        'test_actual': test_data,
        'train_data': train_data,
        'mae': mae,
        'rmse': rmse,
        'model': model
    }

def plot_forecast_results(crime_type, results, last_n_observations=30):
    """
    Plot the forecast results
    """
    plt.figure(figsize=(12, 6))

    # Get the dates for plotting
    train_dates = results['train_data'].index[-last_n_observations:]
    train_values = results['train_data'].values[-last_n_observations:]

    # Plot training data
    plt.plot(train_dates, train_values, label='Historical Data', color='blue')

    # Plot test data
    plt.plot(results['test_actual'].index,
             results['test_actual'].values,
             label='Actual Values',
             color='green')

    # Plot forecast
    plt.plot(results['forecast'].index,
             results['forecast'].values,
             label='Forecast',
             color='red')

    plt.title(f'Crime Forecast for {crime_type}')
    plt.xlabel('Date')
    plt.ylabel('Number of Crimes')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()

    return plt.gcf()

def forecast_all_crime_types(df, top_n=5, forecast_periods=30):
    """
    Forecast for multiple crime types
    """
    # Get top N most frequent crime types
    top_crimes = df['Primary Type'].value_counts().head(top_n).index.tolist()

    forecasts = {}
    for crime_type in top_crimes:
        print(f"Forecasting {crime_type}...")
        forecasts[crime_type] = forecast_crime_type(df, crime_type, forecast_periods)

        # Plot results
        plot_forecast_results(crime_type, forecasts[crime_type])
        plt.show()

        # Print metrics
        print(f"MAE: {forecasts[crime_type]['mae']:.2f}")
        print(f"RMSE: {forecasts[crime_type]['rmse']:.2f}\n")

    return forecasts

# Example usage
if __name__ == "__main__":
    # Ensure date column is datetime
    crimes_2015.loc[:, 'Date'] = pd.to_datetime(crimes_2015.loc[:, 'Date'])

    # Forecast top 5 most common crime types
    forecasts = forecast_all_crime_types(crimes_2015, top_n=5, forecast_periods=30)
```
